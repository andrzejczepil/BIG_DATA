```python
import pandas as pd
df=pd.read_csv("ceny.csv",delimiter=";")
```

```python
df.dtypes
```

```python
df.plot(kind='scatter',x='m2',y='cena')
```

```python
df.plot(kind='scatter',x='pokoje',y='cena')
```

```python
import numpy as np
from sklearn import linear_model
X=np.c_[df['m2']]
y=np.c_[df['cena']]
model=linear_model.LinearRegression()
model.fit(X,y)
print(model.predict([[31]]))
for a in [[40],[50],[60],[70]]:
    print(model.predict([a]))
```

```python
import numpy as np
from sklearn import linear_model
X=np.c_[df['m2'],df['pokoje']]
y=np.c_[df['cena']]
model=linear_model.LinearRegression()
model.fit(X,y)
print(model.predict([[31,1]]))
for a in [[40,2],[50,3],[60,3],[70,5]]:
    print(model.predict([a]))
```
